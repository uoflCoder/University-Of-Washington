```python
import graphlab
```

```python
image_train = graphlab.SFrame('image_train_data/')
image_test = graphlab.SFrame('image_test_data/')
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
image_train['label'].sketch_summary()
```

```python
image_train.show()
```

```python
dog_model = image_train[image_train['label'] == 'dog']
automobile_model = image_train[image_train['label'] == 'automobile']
cat_model = image_train[image_train['label'] == 'cat']
bird_model = image_train[image_train['label'] == 'bird']
```

```python
knn_dog_model = graphlab.nearest_neighbors.create(dog_model,
                                              features=['deep_features'],
                                              label='id')

knn_automobile_model = graphlab.nearest_neighbors.create(automobile_model,
                                              features=['deep_features'],
                                              label='id')

knn_cat_model = graphlab.nearest_neighbors.create(cat_model,
                                              features=['deep_features'],
                                              label='id')

knn_bird_model = graphlab.nearest_neighbors.create(bird_model,
                                              features=['deep_features'],
                                              label='id')
```

```python
knn_cat_model.query(image_test[0:1])
```

```python
def get_images_ids(query_result):
    return image_train.filter_by(query_result['reference_label'], 'id')
```

```python
cat_neighbors = get_images_ids(knn_cat_model.query(image_test[0:1]))
```

```python
cat_neighbors['image'].show()
```

```python
dog_neighbors = get_images_ids(knn_dog_model.query(image_test[0:1]))
```

```python
dog_neighbors['image'].show()
```

```python
deep_learning_model = graphlab.load_model('http://s3.amazonaws.com/GraphLab-Datasets/deeplearning/imagenet_model_iter45')
```

```python
image_train['deep_features'] = deep_learning_model.extract_features(image_train)
```
